```python
from os import stat
# scatter plots
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic
df=pd.read_csv('/content/nhanes_reduced .csv')

# Filter the DataFrame, not just the 'Age' column
filtered_data = df[df['Age'] > 40]

plt.scatter(filtered_data['Age'], filtered_data['DirectChol'],color='red') # Fixed typo: 'colo' to 'color'
plt.title('Age vs DirectChol Leve (Ages>40)')
plt.xlabel('Age(years)')
plt.ylabel('DirectChol')
plt.show()
data.head()
```

```python
#SCATTERED PLOT OF PHYSACTIVEDAYS AND BMI
plt.scatter(filtered_data['BMI'], filtered_data['PhysActiveDays'],color='brown')
plt.title('BMI vs PhysActiveDays')
plt.xlabel('BMI')
plt.ylabel('PhysActiveDays')
plt.show()
data.head()
```

```python
plt.scatter(filtered_data['Age'], filtered_data['Poverty'],color='brown')
plt.title('Age vs Poverty')
plt.xlabel('Age')
plt.ylabel('Poverty')
plt.show()
data.head()
```

```python
#suplots of histogram
data['Work'] = data['Work'].astype('category')
work_categories=data['Work'].cat.categories
fig,axes=plt.subplots(nrows=len('work_categories'),ncols=1,figsize=(10,15),sharex=True)
for ax, category in zip(axes,work_categories):
  sns.histplot(data[data['Work']==category]['BMI'],bins=20 , ax=ax)
  ax.set_title(f'Distribution of BMW for{category}')
plt.tight_layout()
plt.show()
```

```python
# boxplots
sns.boxplot(x='HHIncome',y='BMI',data=data)
plt.title('Boxplots of BMI vs HHIncome')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='Gender',y='BMI',data=data)
plt.title('Boxplots of BMI vs Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='Gender',y='DirectChol',data=data)
plt.title('Boxplots of Gender vs DirectChol')
plt.xticks(rotation=60)
plt.show()
```

```python
# kde plot
plt.figure(figsize=(8,6))
sns.kdeplot(data=data, x='BMI' ,hue='Work',multiple='stack')
plt.title('KDE plot of BMI vs Work')
plt.xlabel('BMI')
plt.ylabel('Density')
plt.show()
```

```python
df=data['BMI']
log_df=np.log(df)
plt.figure(figsize=(10,6))
plt.plot(log_df,label='Log Transformed Data',marker='o',linestyle='-', color='blue')
plt.plot(df,label='Original Data',marker='o',linestyle='-', color='red')

plt.title('Original vs log transformed data')
plt.xlabel('Idex')
plt.ylabel('Value')
```
